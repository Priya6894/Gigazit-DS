```python
from PIL import Image


def pixel_to_coordinates(image_path, pixel_position):
  """
  Converts a pixel location within an image to its corresponding x and y coordinates.

  Args:
      image_path (str): Path to the image file.
      pixel_position (tuple): A tuple containing the pixel coordinates (x, y).

  Returns:
      tuple: A tuple containing the x and y coordinates of the pixel location.

  Raises:
      ValueError: If the pixel position is outside the image boundaries.
      FileNotFoundError: If the image file is not found.
      Exception: For any other unexpected errors.
  """
  try:
    with Image.open(image_path) as img:
      width, height = img.size
      x, y = pixel_position

      if x < 0 or x >= width or y < 0 or y >= height:
        raise ValueError("Pixel position is out of the image boundaries.")

      return x, y
  except FileNotFoundError:
    raise FileNotFoundError(f"Image file '{image_path}' not found.")
  except Exception as e:
    raise Exception(f"An error occurred: {e}")


# Example usage: Convert two different pixel locations to coordinates
image_path = "cat1.jpeg"
display_image_with_pixel(image_path, pixel_position)# Replace with your image path

# Pixel location 1: Left eye (assuming 100, 50)
pixel_position1 = (100, 50)
try:
  x1, y1 = pixel_to_coordinates(image_path, pixel_position1)
  print(f"Pixel location {pixel_position1} corresponds to coordinates: ({x1}, {y1})")
except Exception as e:
  print(e)

# Pixel location 2: Any other point (assuming 250, 180)
pixel_position2 = (250, 180)
try:
  x2, y2 = pixel_to_coordinates(image_path, pixel_position2)
  print(f"Pixel location {pixel_position2} corresponds to coordinates: ({x2}, {y2})")
except Exception as e:
  print(e)
```
